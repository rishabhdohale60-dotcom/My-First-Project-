```python
class Car:
    @staticmethod
    def start():
        print("Car Started..")

    @staticmethod
    def stop():
        print("Car Stooped")

class ToyotaCar(Car):
    def __init__(self,brand):
        self.brand=brand

class Fortuner(ToyotaCar):
    def __init__(self,type):
        self.type = type
        
car1 = Fortuner("diesel")
car1.start()
```

```python
# Super Method
class Car:

    def __init__(self,type):
        self.type=type

    @staticmethod
    def start():
        print("Car Started..")

    @staticmethod  #commeon method for intryer class
    def stop():
        print("Car Stooped")

class ToyotaCar(Car):
    def __init__(self,name,type):
        super().__init__(type)
        self.name=name
        super().start()

car1 = ToyotaCar("prius","electric")
print(car1,type)
```

```python
class Student:
    def __init__(self, Phy, Chem, Math):
        self.Phy = Phy
        self.Chem = Chem
        self.Math = Math
        self.calcpercentage()   # initialize percentage when object is created

    def calcpercentage(self):
        self.percentage = str((self.Phy + self.Chem + self.Math) / 3) + "%"

# Create object
stu1 = Student(94, 84, 88)
print(stu1.percentage)   # prints initial percentage

# Update marks
stu1.Phy = 83
print(stu1.Phy)

# Recalculate percentage
stu1.calcpercentage()
print(stu1.percentage)
```

```python
# abouve using prperty

class Student:
    def __init__(self, Phy, Chem, Math):
        self.Phy = Phy
        self.Chem = Chem
        self.Math = Math

    @property  #Decotator
    def calcpercentage(self):
        return str((self.Phy + self.Chem + self.Math) / 3) + "%"
        
stu1 = Student(94, 84, 88)
print(stu1.calcpercentage)

stu1.Phy = 78
print(stu1.calcpercentage)
```

```python
# Polymorphism :Operator Ocverloading
# When the same operatot do differter work according to

class complex:
    def __init__(self,real ,img):
        self.real = real
        self.img = img 

    def showNumber(self):
        print(self.real,"i +",self.img,"j")
    
    def add(self,num2): #using function
        newReal = self.real + num2.real
        newImg = self.img + num2.img
        return complex(newReal,newImg)
    
num1 = complex(1,4)
num1.showNumber()

num2 = complex(4,6)
num2.showNumber()

num3 = num1.add(num2) #for adding No.1 & No.2
num3.showNumber()
```

```python
# using dander function addition
class complex:
    def __init__(self,real ,img):
        self.real = real
        self.img = img 

    def showNumber(self):
        print(self.real,"i +",self.img,"j")
    
    def __add__(self,num2):  #using daander fun __add__
        newReal = self.real + num2.real
        newImg = self.img + num2.img
        return complex(newReal,newImg)
    
num1 = complex(1,4)
num1.showNumber()

num2 = complex(4,6)
num2.showNumber()

num3 = num1 + num2 #for adding No.1 & No.2
num3.showNumber()
```

```python
# substraing numbers

class complex:
    def __init__(self,real ,img):
        self.real = real
        self.img = img 

    def showNumber(self):
        print(self.real,"i +",self.img,"j")
    
    def __sub__(self,num2):
        newReal = self.real - num2.real
        newImg = self.img - num2.img
        return complex(newReal,newImg)
    
num1 = complex(1,4)
num1.showNumber()

num2 = complex(4,6)
num2.showNumber()

num3 = num1-num2 #for sub No.1 & No.2
num3.showNumber()
```

```python
# precties program
class Circle:
    def __init__(self,radius):
        self.radius = radius

    def Area(self):
        return (22/7) * self.radius ** 2

    def Perimeter(self):
        return 2 * (22/7) * self.radius
    
c1 = Circle(4)
print(c1.Area())
print(c1.Perimeter())
```

```python

class Order:
    def __init__(self, item, price):
        self.item = item
        self.price = price

    def __gt__(self, other):  # only self and other
        return self.price > other.price

odr1 = Order("chips", 60)
odr2 = Order("tea", 30)

print(odr1 > odr2)  # True
```
