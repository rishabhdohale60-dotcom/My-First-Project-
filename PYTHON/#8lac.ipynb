```python
#Methods

class student:
    college_name ="netaji"
    
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
        
    def welcome(self):
        print("welcome student,",self.name)

    def get_marks(self):
        return self.marks
    
s1 = student("rishi",98)
s1.welcome()
print(s1.get_marks())
```

```python
#pratice program

class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks= marks

    def get__avg(self):
        total = 0 
        for i in self.marks:  #itrating over the number

            total += i
            avg = total / len(self.marks)
        print("Hi",self.name,"your avg score is:",avg)

s1 = Student("rishabh", [98,76,87] )
s1.get__avg()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the line x + y = 0
x = np.linspace(-5, 5, 400)
y = -x  # boundary line

# Plot the half-plane (x + y <= 0)
plt.fill_between(x, y, -10, alpha=0.3, color='lightblue', label='x + y <= 0')

# Plot the boundary line
plt.plot(x, y, 'b-', label='x + y = 0')

# Plot the point (0,0) from x^2 + y^2 = 0
plt.plot(0, 0, 'ro', markersize=8, label='x^2 + y^2 = 0')

# Formatting
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of x + y <= 0 and x^2 + y^2 = 0')
plt.show()

```
