```python
#Rounding Decimal
import numpy as np

#truncation 
arr = np.trunc([-2.122,-1.212]) # closet to Zero
print(arr)


#Fix
arr = np.fix([-1.2233,-2.123])
print(arr)


#rounding ----around():
arr = np.around(3.1666, 2)  #simple rounding that if 1 => 5 else do nothing
print(arr)  

#Floor
arr = np.floor([-3.5455,5.123])  #nearest lower decimal
print(arr)

#ceil
arr = np.ceil([-3.5455,5.123])  #nearest upper decimal
print(arr)

```

```python
# numpy summation

import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4,5,6])

newarr = np.add(arr1, arr2)
print(newarr)

arr1 = np.array([1, 2, 3])
arr2 = np.array([7,4,8])

newarr1 = np.sum([arr1, arr2])
print(newarr1)

arr1 = np.array([1, 2, 3])
arr2 = np.array([9,8,7,])

newarr2 = np.sum([arr1, arr2],axis = 1)
print(newarr2)


arr = np.array([1,2,3])

newarr3 =np.cumsum(arr)
print(newarr3)

```

```python
# numpy product

import numpy as np

arr1 = np.array([1,2,3,4])
x = np.prod(arr1)
print(x)


arr1 = np.array([12,2,3])
arr2 = np.array([13,4,2])
y = np.prod([arr1,arr2])
print(y)


arr1 = np.array([12,2,3])
arr2 = np.array([13,4,2])
y = np.prod([arr1,arr2], axis = 1)
print(y)

# commutative product

arr1 = np.array([1,2,3,4])
x = np.cumprod(arr1)
print(x)


arr1 = np.array([12,2,3])
arr2 = np.array([13,4,2])
y = np.cumprod([arr1,arr2])
print(y)
```
